Extract text from resume

```python
def extract_text(pdfpath):
  reader=PdfReader(pdfpath)
  text=""
  for p in reader.pages:
    if p.extract_text():
      text+=p.extract_text()
  return text.strip()
resume_text=extract_text("resume.pdf")
#project manager
print ("resume:=>", resume_text[:300])#print fist 300 elements


```

Job Description

```python
job_description=""" A project manager is responsible for planning, executing, and closing projects on time, within budget, and according to scope.
Key duties include defining project goals, managing resources, leading the project team, monitoring progress, managing risks, and communicating with stakeholders.
This role requires a blend of planning, organizational, and leadership skills to ensure a project's success and align with company strategy.
 """
```

Embedding ( capture sematic meaning)

```python
embedder=OpenAIEmbeddings()
resume_vector=embedder.embed_query(resume_text)
job_desc_vector=embedder.embed_query(job_description)
print( "resume_vector=>", resume_vector[:3])#print fist 10 elements
print( "job_desc_vector=>", job_desc_vector[:3])#print fist 10 elements
```

calculate cosine similarity

```python
score=cosine_similarity([resume_vector],[job_desc_vector])[0][0]
print(f' Match Score=>" {score: .2f}')
```
